```python
# Imports
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Let's now create our data 

def create_chart():
    x_series = np.arange(20)
    y_series = x_series ** 2

    # Plotting 
    plt.plot(x_series, y_series)
    # Title of the canvas
    plt.title('Sample Chart')
    # X-axis label
    plt.xlabel('X Axis')
    # Y-axis label
    plt.ylabel('Y Axis')

    # Printing the canvas
    plt.show();
```

```python
create_chart()
```

```python
# Now I will practise visualizations using the weather data.
import pandas as pd 

weather_df = pd.read_csv('weather_2012.csv', parse_dates=True, index_col='Date/Time')
weather_df.head(5)
```

```python
# Let's generate a line chart that visualizes the temparature readings in the month of January.

# We first collect the respective months from the Date/Time column
monthly_df = weather_df.groupby(weather_df.index.month).mean()
monthly_df.head(25)
```

```python
x_series = monthly_df.index
y_series = monthly_df['Temp (C)']
```

```python
# Let's now plot the line graph

plt.plot(x_series, y_series)
#T itle
plt.title('Monthly Temperature trend in 2012')
# X-axis label
plt.xlabel('Month')
# Y-axis label
plt.ylabel('Temperature (C)')
# Print the graph
plt.show()
```

```python
# Let's now change the X-axis labels to the respective months 
import calendar
calendar_months = calendar.month_name[1:]

print(calendar_months)
```

```python
x = calendar_months
y = monthly_df['Temp (C)']
plt.plot(x,y)
plt.title('Monthly Temperature Trend, 2012')
plt.xlabel('Month')
plt.ylabel('Temp (C)')
plt.show()
```

```python
# Let's now find the distribution of values in the Wind Spd column

def histogram_plot():
    plt.hist(weather_df['Wind Spd (km/h)'])
    plt.title('Distribution of Wind Speed (km/h)')
    plt.xlabel('Wind Speed (km/h)')
    plt.ylabel('Frequency');

histogram_plot()
```

```python
# Let's create a bar plot to show the Visibility by week.

def bar_plot():
    weekly_data = weather_df.groupby(weather_df.index.dayofweek).mean()
    plt.bar(weekly_data.index, weekly_data['Visibility (km)'])

    plt.title('Visibility by week, 2012')
    plt.xlabel('Day of week')
    plt.ylabel('Visibility (km)')

    plt.xticks(weekly_data.index, calendar.day_abbr, rotation=45)

    plt.show()
    
bar_plot()
```

```python
# Let's now create boxplots from sample data

x = np.arange(0, 10, 0.1)
y = np.exp(x)

plt.boxplot(y)
plt.show()
```

```python
# Let's look if the Temperature and Pressure are correlated in the month of January.

jan_df = weather_df['2012-01']

def scatter_plot():
    plt.scatter(x=jan_df['Temp (C)'],y=jan_df['Stn Press (kPa)'])   
    plt.xlabel('Temp (C)')
    plt.ylabel('Pressure')
    plt.title('Corelation between Temperature and Pressure for January')
    
scatter_plot()
```
