```python

# Imports
import pandas as pd
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt


from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import KMeans
```

```python
# Loading our dataset
df= pd.read_csv('diabetes.csv')
df.head()
```

```python
# Let's now drop the diabetes column in the above dataset and store it in a variable 
# called **y** and fit a `KMeans()` with `n_clusters=2`

# Stored the diabetes column in a variable y
y = df.iloc[:,-1] 
new_df = df.drop(['diabetes'],axis=1)
new_df.head()

kmeans = KMeans(n_clusters=2)
kmeans.fit(new_df)
```

```python
# Let's now find the total number of samples that the model misclassified.

y = np.array(y)
mis_classified = np.where(kmeans.labels_ != y)
len(mis_classified[0])
```

```python
# Now we'll plot a Dendrogram for the entire dataset.

plt.figure(figsize=(10, 7))
plt.title("Customer Dendrogram")

dend = sch.dendrogram(sch.linkage(new_df, method='ward'),leaf_rotation=90,leaf_font_size=2.4)
plt.show()
```
