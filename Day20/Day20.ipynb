# Linear Regression exercise

```python
# Imports
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Import the dataset
df = pd.read_csv('Advertising.csv')
df.head(10)
```

### Find the relationship between the predictor and the target variables.


```python
plt.figure(figsize=(20,5))
plt.subplot(131)
plt.scatter(df['TV'],df['sales'])
plt.xlabel("Advertising on TV")
plt.ylabel("Sales")
plt.subplot(132)
plt.scatter(df['radio'],df['sales'])
plt.xlabel('Advertising on Radio')
plt.ylabel('Sales')
plt.subplot(133)
plt.scatter(df['newspaper'],df['sales'])
plt.xlabel('Advertising on Newspaper')
plt.ylabel('Sales');
```

### Slice the predictors and target into variables X and y respectively.

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Let's now fit a Linear regression model on the variable X.
from sklearn.linear_model import LinearRegression

model =LinearRegression()
model.fit(X,y)
```

```python
# Let's find the intercept and coefficients for the above model
print("Coefficients: ", model.coef_)
print("Intercept: ", model.intercept_)
```

```python
# Let's noe try predicting the variable X and create a Dataframe of the actual values.
y_pred = model.predict(X)
result = pd.DataFrame({'Actual':y, 'Predicted': y_pred})
result.head(10)
```

```python
print("The Mean Square Error is:", mean_squared_error(y_pred, y))
```
