```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('weather_2012.csv')
data.tail(10)
```

```python
# Mean and median for Wind Speed and Temperature columns
wind_speed_mean = data['Wind Spd (km/h)'].mean()
wind_speed_median = data['Wind Spd (km/h)'].median()

temp_mean = data['Temp (C)'].mean()
temp_median = data['Temp (C)'].median()

print("Wind Speed => Mean: ",wind_speed_mean,end='   ')
print("Median: ",wind_speed_median)

print("Temperature => Mean: ",temp_mean,end='   ')
print("Median: ",temp_median)
```

```python
# Mean and the IQR for our Relative Humidity column
humidity = data['Rel Hum (%)']
humidity_median = np.median(data['Rel Hum (%)'])
print("Humidity Median: ",humidity_median)

q1 = humidity.quantile(0.25)
q3 = humidity.quantile(0.75)
print('Q1:', q1)
print('Q3:', q3)
print('IQR:', q3 - q1)
```

```python
# Range of the Relative Humidity column
range = np.max(humidity) - np.min(humidity)
range
```

```python
# Mode of the Weather column
weather_mode = data['Weather'].mode()
print("Mode for weather : ",weather_mode[0])
```

```python
# Skewness for the Temperature and Wind Speed columns
plt.figure(figsize=[10,5],)

plt.subplot(1,2,1)
plt.title('Wind Speed Plot')
plt.hist(data['Wind Spd (km/h)'])
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Frequency')
plt.subplot(1,2,2)
plt.title('Temperature Plot')
plt.hist(data['Temp (C)'])
plt.xlabel("Temperature");
```

```python
# Variance and standard deviation of the Dew Point Temperature column
variance = data['Dew Point Temp (C)'].var()
standard_deviation = np.sqrt(variance)
print("Variance: ",variance)
print("Standard deviation: ",standard_deviation)
```

```python
# Correlation between all continuous features in the dataset
correlation = data.corr()
sns.heatmap(correlation);
```
